```python
usa = df[df.Country == 'USA']
usa
```

```python
usa_winners = usa.groupby(['Year','Athlete','Medal']).size().unstack(fill_value=0)
usa_winners['Total']=usa_winners['Gold']+ usa_winners['Silver']+usa_winners['Bronze']
usa_winners
```

```python
usa_winners = usa_winners.reset_index()[['Year','Athlete','Total']]
year = []
athlete = []
medal = []
for y in usa_winners.Year.unique():
    new_df = usa_winners[usa_winners.Year == y]
    y_ath_m = new_df.sort_values('Total',ascending=False).iloc[0]
    year.append(y_ath_m[0])
    athlete.append(y_ath_m[1])
    medal.append(y_ath_m[2])
usa_top_winners = pd.DataFrame({'Year':year, 'Athlete':athlete, 'Total':medal})
usa_top_winners
```

```python
usa_top_winners.info()
```

```python
usa_top_winners['Year']=usa_top_winners['Year'].astype(str)
usa_top_winners['Brief']=usa_top_winners['Athlete']+ ' / ' +usa_top_winners['Year']
usa_top_winners
```
